```python
def drop_non_varying_column(dataframe):
    # Dockstring
    '''  
    Removes the columns in the dataframe that contain the constant numbers.
    
     Args:
            dataframe     : The input dataframe to be proccessed
            
    Reurns:
            dataframe: contains the columns with valiable numarical values
    '''
    
    describe=dataframe.describe()
    # Created dataframe with describe value
    colums=[]
    for label in describe:
        if describe.loc['max'][label]==describe.loc['min'][label]:
            colums.append(label)
    # Listing out the columns where data remains unchanged throught the experiment 
    
    dataframe.drop(colums,axis=1,inplace=True)
    # Droping the listed columns
    
    return (dataframe) 
```

```python
def no_of_cycle_left(dataframe):
    
    '''
        Creates the new column representing no of cycle left before the failure of engine
    
     Args:
            dataframe     : The input dataframe to be proccessed
            
    Reurns:
            dataframe: contains the column with no of cycle left
            
    
    '''
    # Finding engine id value count
    freq=dataframe['id'].value_counts()
    
    # Creating the new column 'Max No Cycle' and mapping it to frequency
    dataframe['Max_no_cycle']=dataframe['id']
    dataframe['Max_no_cycle']=dataframe.Max_no_cycle.map(freq)
    
    # Creacting the new column "No of cycle left" = Max No of cycle- cycle
    dataframe['No_of_cycle_left']=dataframe['Max_no_cycle']-dataframe['cycle']
    
    #droping the columns ['id','Max_no_cycle']
    dataframe = dataframe.drop(['id','Max_no_cycle'],axis=1)
    
    return dataframe
```

```python
df=drop_non_varying_column(df)
df=no_of_cycle_left(df)
```
